```python
df['total_runs'] = df['runs_off_bat'] + df['wides'] + df['noballs']
```

```python
innings_totals = df.groupby(['match_id', 'innings'])['total_runs'].sum().reset_index()
```

```python
match_results = innings_totals.pivot(index='match_id', columns='innings', values='total_runs').reset_index()
match_results.columns = ['match_id', 'team1_runs', 'team2_runs']
```

```python
match_results['Win_by_runs'] = match_results.apply(
    lambda x: x['team1_runs'] - x['team2_runs'] if x['team1_runs'] > x['team2_runs'] else None, axis=1)
match_results['Win_by_wickets'] = match_results.apply(
    lambda x: 10 if x['team2_runs'] > x['team1_runs'] else None, axis=1)
```
