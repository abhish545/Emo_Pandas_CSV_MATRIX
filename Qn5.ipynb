```python
dl= pd.read_excel('C:\\Users\\MY\\Downloads\\deleveries.xlsx',header=3)
dl.head()
```

```python
dl.head().drop('Unnamed: 0',axis=1)
```

```python
# Convert columns to numeric where needed to ensure calculations work properly
# - 'ball' should be float to handle ball-by-ball data correctly.
# - 'runs_off_bat', 'wides', and 'noballs' are converted to numeric values to calculate total runs.
dl['ball'] = dl['ball'].astype(float)
```

```python
dl['runs_off_bat'] = pd.to_numeric(dl['runs_off_bat'],errors='coerce')
```

```python
dl['wides'] = pd.to_numeric(dl['wides'],errors='coerce').fillna(0)
dl['noballs']= pd.to_numeric(dl['noballs'],errors='coerce').fillna(0)
```

```python
# Calculate total runs scored in each ball by adding runs off the bat, wides, and no-balls.
# This helps in getting the actual runs for each delivery, especially important for overs 41-50.
dl['total_runs']= dl['runs_off_bat'] + dl['wides'] + dl['noballs']
```

```python
# Extract the over number from the 'ball' column (using int conversion)
# and filter to get only deliveries in the third powerplay (overs 41-50).
dl['overs']= dl['ball'].apply(lambda x:int(x))
third_powerplay_dl = dl[(dl['overs']>=41)&(dl['overs']<=50)]
```

```python
# Define a column 'legal_delivery' to identify legal balls (balls without wides or no-balls).
# This is used to accurately count legal deliveries in each over
third_powerplay_dl['legal_delivery'] = (third_powerplay_dl['wides'] == 0) & (third_powerplay_dl['noballs'] == 0)
```

```python
# Group by each team and innings to calculate the total runs scored and total legal deliveries faced in the third powerplay.
# This summary provides the basis for calculating run rates in the third powerplay for each team.
summary_df = third_powerplay_dl.groupby(['batting_team', 'innings']).agg(
    Total_Runs_3rd_Powerplay=('total_runs', 'sum'),
    Total_Deliveries_3rd_Powerplay=('legal_delivery', 'sum')  # Only count legal deliveries
).reset_index()
```

```python
# Calculate the run rate for the third powerplay for each team and innings
# Run Rate = (Total Runs Scored) / (Total Overs Played) in the third powerplay
# Total overs is derived by dividing legal deliveries by 6.
summary_df['Run_Rate_3rd_Powerplay'] = (
    summary_df['Total_Runs_3rd_Powerplay'] / (summary_df['Total_Deliveries_3rd_Powerplay'] / 6))
```

```python
# Calculate the aggregate run rate across both innings for each team in the third powerplay
# This step combines the run rate data for both innings, giving an overall picture for each team.
# Separate the run rates for first and second innings for each team to analyze differences
# The resulting data frames contain run rates for the third powerplay when teams batted first or second.
aggregate_run_rate = summary_df.groupby('batting_team').agg(
    Aggregate_Run_Rate_3rd_PP=('Total_Runs_3rd_Powerplay', 'sum'),  # Total runs across both innings
    Total_Deliveries=('Total_Deliveries_3rd_Powerplay', 'sum')      # Total legal deliveries across both innings
).reset_index()
aggregate_run_rate['Aggregate_Run_Rate_3rd_PP'] = aggregate_run_rate['Aggregate_Run_Rate_3rd_PP'] / (aggregate_run_rate['Total_Deliveries'] / 6)
first_innings_run_rate = summary_df[summary_df['innings'] == 1][['batting_team', 'Run_Rate_3rd_Powerplay']].rename(
    columns={'Run_Rate_3rd_Powerplay': 'Run_Rate_3rd_Powerplay_Batting_First'}
)
second_innings_run_rate = summary_df[summary_df['innings'] == 2][['batting_team', 'Run_Rate_3rd_Powerplay']].rename(
    columns={'Run_Rate_3rd_Powerplay': 'Run_Rate_3rd_Powerplay_Batting_Second'}
)
# Merge all run rate data to have a comprehensive view of each team’s performance in the third powerplay
# Includes aggregate, first innings, and second innings run rates.
run_rate_comparison = pd.merge(first_innings_run_rate, second_innings_run_rate, on='batting_team', how='outer')
```

```python
# Calculate the number of wickets taken by each team in the last powerplay and the average deliveries per wicket.
# Group by the bowling team, count the total wickets, and sum legal deliveries in the third powerplay for each bowling team.
wickets_summary = third_powerplay_dl[third_powerplay_dl['wicket_type'].notna()].groupby('bowling_team').agg(
    Total_Wickets_3rd_Powerplay=('wicket_type', 'count'),
    Total_Deliveries_Bowled_3rd_Powerplay=('legal_delivery', 'sum')  # Only count legal deliveries
).reset_index()

# Calculate the average number of deliveries per wicket taken in the last powerplay
# This helps understand the bowling team’s efficiency in taking wickets during this phase.
wickets_summary['Avg_Deliveries_per_Wicket'] = (
    wickets_summary['Total_Deliveries_Bowled_3rd_Powerplay'] / wickets_summary['Total_Wickets_3rd_Powerplay']
)
```

```python
# Merge the wickets data with run rate data to form the final summary table
final_summary = run_rate_comparison.merge(wickets_summary, left_on='batting_team', right_on='bowling_team', how='left').drop(columns='bowling_team')

# Rename columns for final presentation
final_summary.columns = ['Team', 'Aggregate Run Rate (3rd PP)', 'First Innings Run Rate (3rd PP)', 'Second Innings Run Rate (3rd PP)', 'Wickets Taken (Last PP)', 'Avg Deliveries per Wicket (Last PP)']

# Display the final summary table
print(final_summary)
```
