```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('CSV_Matrix_2.0_Group_B_dataset.xlsx', sheet_name='Deliveries', header=3)
```

```python
df = df.drop('Unnamed: 0', axis=1)
df = df.fillna(0)
df.head()
```

```python
new_col = ['Team', 'Played', 'Won', 'Lost', 'NRR', 'NRR For', 'NRR Against', 'Extras', '%total runs given in extras']
ndf = pd.DataFrame(columns= new_col)
team = ['India', 'South Africa', 'Australia', 'New Zealand', 'Pakistan', 'Afghanistan', 'England', 'Bangladesh', 'Sri Lanka', 'Netherlands']
ndf['Team']= team
ndf
```

```python
def extras(x):
    extras= 0
    for index, row in df.iterrows():
        if row['batting_team'] == x:
            extras= extras+ row['wides']+ row['noballs']+row['byes']+row['legbyes']+row['penalty']
    return extras

team = ['India', 'South Africa', 'Australia', 'New Zealand', 'Pakistan', 'Afghanistan', 'England', 'Bangladesh', 'Sri Lanka', 'Netherlands']
i=0
ind=0
for i in team:
    ans = 0;
    ans= extras(i)
    ndf.iloc[ind,7] = ans
    ind=ind+1
ndf
```

```python
df['total_runs'] = df[['wides' , 'noballs' , 'byes' , 'legbyes' , 'penalty' , 'runs_off_bat']].sum(axis=1)
team_score = df.groupby(['match_id' , 'batting_team'])['total_runs'].sum().reset_index()
winner = team_score.loc[team_score.groupby('match_id')['total_runs'].idxmax()]
winner= winner.rename(columns={'batting_team': 'winners'})
winner.head(3)
```

```python
def winners(x):
    count= 0
    for index, row in winner.iterrows():
        if row['winners'] == x:
            count=count+1
    return count

team = ['India', 'South Africa', 'Australia', 'New Zealand', 'Pakistan', 'Afghanistan', 'England', 'Bangladesh', 'Sri Lanka', 'Netherlands']
i=0
ind=0
for i in team:
    ans = 0;
    ans= winners(i)
    ndf.iloc[ind,2] = ans
    ind=ind+1
ndf
```

```python
looser = team_score.loc[team_score.groupby('match_id')['total_runs'].idxmin()]
looser= looser.rename(columns={'batting_team': 'loosers'})
looser.head(3)
```

```python
def lsers(x):
    count= 0
    for index, row in looser.iterrows():
        if row['loosers'] == x:
            count=count+1
    return count

team = ['India', 'South Africa', 'Australia', 'New Zealand', 'Pakistan', 'Afghanistan', 'England', 'Bangladesh', 'Sri Lanka', 'Netherlands']
i=0
ind=0
for i in team:
    ans = 0;
    ans= lsers(i)
    ndf.iloc[ind,3] = ans
    ind=ind+1
ndf
```

```python
team = ['India', 'South Africa', 'Australia', 'New Zealand', 'Pakistan', 'Afghanistan', 'England', 'Bangladesh', 'Sri Lanka', 'Netherlands']
i=0
ind=0
for i in team:
    ndf.iloc[ind,1] = ndf.iloc[ind,2]+ndf.iloc[ind,3] 
    ind=ind+1
ndf
```

```python
def tot(x):
    runs,oversfaced,runscon,oversbowled = 0,0,0,0
    for index, row in df.iterrows():
        if row['batting_team'] == x:
            runs= runs+ row['wides']+ row['noballs']+row['byes']+row['legbyes']+row['penalty']+row['runs_off_bat']
            oversfaced = oversfaced+ row['innings']
        if row['bowling_team'] == x:
            runscon= runscon+ row['wides']+ row['noballs']+row['byes']+row['legbyes']+row['penalty']+row['runs_off_bat']
            oversbowled= oversbowled+ row['innings']
    return runs,oversfaced/6,runscon,oversbowled/6
    

team = ['India', 'South Africa', 'Australia', 'New Zealand', 'Pakistan', 'Afghanistan', 'England', 'Bangladesh', 'Sri Lanka', 'Netherlands']
i=0
ind =0
for i in team:
    truns,ofaced,rcon,obow = 0,0,0,0
    nrr,nrrfor,nrragainst=0,0,0
    truns,ofaced,rcon,obow = tot(i)
    nrr= (truns/ofaced) - (rcon/obow)
    nrrfor= (truns/ofaced)
    nrragainst= (rcon/obow)
    per_extras = (ndf.iloc[ind,7]/truns)*100
    ndf.iloc[ind,4] = nrr
    ndf.iloc[ind,5] = nrrfor
    ndf.iloc[ind,6] = nrragainst
    ndf.iloc[ind,8] = per_extras
    ind=ind+1
    
ndf
    
```
